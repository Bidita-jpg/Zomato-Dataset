```python
#about MISSING VALUE
df.isnull().sum()
# as we can see here only 9 null values for cuisines
```

```python
#by using LIST COMPREHENSION find missing values---here cuisines is null
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
#by plottingheatmap missing value
plt.rcParams["figure.figsize"]=(15,6)
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
df_country=pd.read_excel('Country-code.xlsx')#here one more dataset loading
df_country.head()
```

```python
df.columns
```

```python
##combine the df(country code) and df_country(country code)?


#Method-1

pd.merge(df,df_country)
```

```python
#Method-2
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
#from here we can see that the maximum no of ZOMATO is PRESENT in INDIA
```

```python
#only see country name
country_name=final_df.Country.value_counts().index
country_name
```

```python
#only see country counts
country_value=final_df.Country.value_counts().values
country_value
```

```python
##PIE CHART top 3 countries using zomato


plt.pie(country_value[:3],labels=country_name[:3],colors='by',autopct='%1.1f%%')
```

# OBSERVATION-
Zomato maximum records or transactions are from INDIA,and then USA and then UNITED KINGDOM.

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])#we want to know overall rating 
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

```python
plt.rcParams["figure.figsize"]=(25,10)
sns.barplot(x="Aggregate rating",y="Rating count",data=rating)
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(25,10)
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=rating, palette=['blue','red','orange','yellow','green'])
plt.show()
```
